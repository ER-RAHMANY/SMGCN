## Over-Smoothing

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.datasets import Planetoid
import numpy as np
import matplotlib.pyplot as plt


dataset_name = "Cora"  
dataset = Planetoid(root=f'./data/{dataset_name}', name=dataset_name)
data = dataset[0]

# Définition du modèle GCN
class GCN(torch.nn.Module):
    def __init__(self, num_layers, in_channels, hidden_channels, out_channels):
        super(GCN, self).__init__()
        self.num_layers = num_layers
        self.convs = torch.nn.ModuleList()
        self.convs.append(GCNConv(in_channels, hidden_channels))
        for _ in range(num_layers - 2):
            self.convs.append(GCNConv(hidden_channels, hidden_channels))
        self.convs.append(GCNConv(hidden_channels, out_channels))
    
    def forward(self, x, edge_index):
        for conv in self.convs:
            x = conv(x, edge_index)
            x = F.relu(x)
        return x

# Fonction pour mesurer la variance intra-classe
def compute_variance(embeddings, labels):
    variances = []
    for label in labels.unique():
        class_embeddings = embeddings[labels == label]
        if len(class_embeddings) > 1:
            var = class_embeddings.var(dim=0).mean().item()
            variances.append(var)
    return np.mean(variances)

# Tester différentes profondeurs de GCN
layer_depths = [2, 4, 8, 16]
variances = []

for depth in layer_depths:
    model = GCN(num_layers=depth, in_channels=dataset.num_node_features, 
                hidden_channels=16, out_channels=dataset.num_classes)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
    
    model.train()
    for epoch in range(200):  # Entraînement sur 200 epochs
        optimizer.zero_grad()
        out = model(data.x, data.edge_index)
        loss = F.cross_entropy(out[data.train_mask], data.y[data.train_mask])
        loss.backward()
        optimizer.step()
    
    model.eval()
    with torch.no_grad():
        embeddings = model(data.x, data.edge_index)
        variance = compute_variance(embeddings, data.y)
        variances.append(variance)
        print(f"Profondeur {depth} : Variance intra-classe = {variance:.6f}")

# Affichage des résultats
plt.figure(figsize=(8,5))
plt.plot(layer_depths, variances, marker='o', linestyle='-')
plt.xlabel("Nombre de couches GCN")
plt.ylabel("Variance intra-classe des embeddings")
plt.title(f"Sur-lissage du GCN sur {dataset_name}")
plt.show()
```
