## Further Qualitative Analysis


```python
import time
import torch
import torch.nn.functional as F
from torch_geometric.datasets import Planetoid
from torch_geometric.nn import GCNConv
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE
```

###  Visualizing Hidden Layer Embeddings (t-SNE)

We train a twolayer GCN on Cora, extract the activations from the hidden layer, and use tSNE to project them to 2D so that we can visually inspect how well the classes are separated. This is useful for studying oversmoothing effects..

```python
dataset_name = "Cora"
dataset = Planetoid(root='./data/Cora', name=dataset_name)
data = dataset[0]

#a GCN model with variable depth, returning the last hidden activations
class GCN_Activations(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, num_hidden_layers=1):
        """
        Args:
            in_channels:     Number of input features
            hidden_channels: Number of hidden units per layer
            out_channels:    Number of output classes
            num_hidden_layers: How many hidden layers (each followed by ReLU)
                               before the final classification layer.
                               Example: If num_hidden_layers=1, then we have:
                                 conv1 -> ReLU -> conv_final
        """
        super(GCN_Activations, self).__init__()
        self.convs = torch.nn.ModuleList()
        
        self.convs.append(GCNConv(in_channels, hidden_channels))
        
        for _ in range(num_hidden_layers - 1):
            self.convs.append(GCNConv(hidden_channels, hidden_channels))
        
        self.final_conv = GCNConv(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        """
        Returns:
            hidden: the output of the last hidden layer (after ReLU)
            out:    the output logits of the final classification layer
        """
        for conv in self.convs:
            x = conv(x, edge_index)
            x = F.relu(x)
        hidden = x  # Last hidden representation
        out = self.final_conv(hidden, edge_index)
        return hidden, out


def train_model(model, data, epochs=200, lr=0.01, weight_decay=5e-4):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    model.train()
    for epoch in range(epochs):
        optimizer.zero_grad()
        hidden, out = model(data.x, data.edge_index)
        loss = F.cross_entropy(out[data.train_mask], data.y[data.train_mask])
        loss.backward()
        optimizer.step()

#experiment with different GCN depths & visualize t-SNE
depths = [2, 4, 8, 16]  #total layers = #hidden layers + 1 final layer
tsne_results = {}      

for d in depths:
    model = GCN_Activations(
        in_channels=dataset.num_node_features,
        hidden_channels=32,
        out_channels=dataset.num_classes,
        num_hidden_layers=d - 1
    )
    
    train_model(model, data, epochs=200, lr=0.01, weight_decay=5e-4)
    
    model.eval()
    with torch.no_grad():
        hidden, out = model(data.x, data.edge_index)
    
    #performing t-SNE on the hidden representations
    hidden_np = hidden.cpu().numpy()
    tsne = TSNE(n_components=2, random_state=42)
    hidden_2d = tsne.fit_transform(hidden_np)
    
    tsne_results[d] = hidden_2d

#plotting t-SNE for each depth side by side
num_plots = len(depths)
plt.figure(figsize=(4 * num_plots, 4))
for i, d in enumerate(depths, 1):
    plt.subplot(1, num_plots, i)
    scatter = plt.scatter(
        tsne_results[d][:, 0],
        tsne_results[d][:, 1],
        c=data.y.cpu().numpy(),
        cmap='Set1',
        s=10
    )
    plt.title(f"t-SNE of Hidden Activations (Depth={d})")
    plt.xlabel("Dim 1")
    plt.ylabel("Dim 2")
    
plt.tight_layout()
plt.show()
```

#### -> At lower depths (2 and 4..), the plots show distinct clusters -> the GCN effectively captures class-specific features early on.. but at depths 8 and especially 16, we see more overlap and less clear separation, which indicates over-smoothing (and it is a common issue in deeper GCNs..)
#### -> This blending shows how node embeddings converge as each node aggregates information from larger neighborhoods, which ultimately dilutes discriminative power..

## Robustness to Low Label Rates

We vary the percentage of training nodes per class (from 5% to 50%) and then evaluates the models test accuracy on CiteSeer. The results will help analyze how the model behaves when labels are extremely sparse.

```python

dataset_name = "CiteSeer"
dataset = Planetoid(root='./data/CiteSeer', name=dataset_name)
data = dataset[0]

def get_random_train_mask(data, fraction=0.1):
    #create a new train_mask with a given fraction of nodes per class
    train_mask = torch.zeros(data.num_nodes, dtype=torch.bool)
    for c in data.y.unique():
        idx = (data.y == c).nonzero(as_tuple=False).view(-1)
        n_train = max(1, int(fraction * idx.size(0)))
        selected = idx[torch.randperm(idx.size(0))[:n_train]]
        train_mask[selected] = True
    return train_mask

#a simple 2-layer GCN (reuse from before)
class GCN(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, out_channels)
    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x

fractions = [0.05, 0.1, 0.2, 0.3, 0.5]
test_accuracies = []

for frac in fractions:
    print(f"Training with {frac*100:.0f}% of nodes per class")
    #create a new training mask with the given fraction per class
    data.train_mask = get_random_train_mask(data, fraction=frac)
    
    model = GCN(dataset.num_node_features, 32, dataset.num_classes)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
    
    model.train()
    for epoch in range(300):
        optimizer.zero_grad()
        out = model(data.x, data.edge_index)
        loss = F.cross_entropy(out[data.train_mask], data.y[data.train_mask])
        loss.backward()
        optimizer.step()
    
    model.eval()
    with torch.no_grad():
        out = model(data.x, data.edge_index)
        pred = out.argmax(dim=1)
        correct = pred[data.test_mask] == data.y[data.test_mask]
        acc = int(correct.sum()) / int(data.test_mask.sum())
        test_accuracies.append(acc)
        print(f"Test Accuracy: {acc:.4f}")

plt.figure(figsize=(8,6))
plt.plot([f*100 for f in fractions], test_accuracies, marker='o')
plt.xlabel("Percentage of training nodes per class (%)")
plt.ylabel("Test Accuracy")
plt.title("Robustness to Low Label Rates on CiteSeer")
plt.show()
```

#### -> Results on CiteSeer (where accuracy steadily improves as the percentage of labeled nodes increases) nicely illustrate the semi-supervised strength of GNNs. The jump from ~0.73 accuracy at 5% labels to ~0.86 accuracy at 50% labels is quite significant and expected..
#### -> might be interesting to include confidence intervals or standard deviations across multiple runs with different random seeds to show the consistency (or variance) of the method at each label rate..

## Sensitivity / Ablation Study - Varying GCN Depth

We loop over several choices for the number of layers (2, 4, 8, 16) on the Cora dataset. For each depth, we train the network, compute the test accuracy, and measure the intraclass variance (a proxy for oversmoothing). The plots show the effect on both accuracy and hidden representation variance..

```python

dataset_name = "Cora"
dataset = Planetoid(root=f'./data/{dataset_name}', name=dataset_name)
data = dataset[0]

#a multilayer GCN that also returns the final hidden layer activations.
class GCN_Activations(torch.nn.Module):
    def __init__(self, num_layers, in_channels, hidden_channels, out_channels):
        super(GCN_Activations, self).__init__()
        self.num_layers = num_layers
        self.convs = torch.nn.ModuleList()
        self.convs.append(GCNConv(in_channels, hidden_channels))
        for _ in range(num_layers - 2):
            self.convs.append(GCNConv(hidden_channels, hidden_channels))
        self.convs.append(GCNConv(hidden_channels, out_channels))
    def forward(self, x, edge_index):
        for conv in self.convs[:-1]:
            x = conv(x, edge_index)
            x = F.relu(x)
        hidden = x.clone()  #hidden layer before final layer
        x = self.convs[-1](x, edge_index)
        return hidden, x

def compute_intra_class_variance(embeddings, labels):
    variances = []
    for c in labels.unique():
        class_emb = embeddings[labels == c]
        if class_emb.size(0) > 1:
            var = class_emb.var(dim=0).mean().item()
            variances.append(var)
    return np.mean(variances)

depths = [2, 4, 8, 16]
test_accs = []
intra_class_vars = []

for depth in depths:
    model = GCN_Activations(num_layers=depth, 
                            in_channels=dataset.num_node_features,
                            hidden_channels=32, 
                            out_channels=dataset.num_classes)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
    model.train()
    for epoch in range(300):
        optimizer.zero_grad()
        hidden, out = model(data.x, data.edge_index)
        loss = F.cross_entropy(out[data.train_mask], data.y[data.train_mask])
        loss.backward()
        optimizer.step()
    model.eval()
    with torch.no_grad():
        hidden, out = model(data.x, data.edge_index)
        pred = out.argmax(dim=1)
        correct = pred[data.test_mask] == data.y[data.test_mask]
        acc = int(correct.sum()) / int(data.test_mask.sum())
        test_accs.append(acc)
        var = compute_intra_class_variance(hidden, data.y)
        intra_class_vars.append(var)
        print(f"Depth {depth} - Test Accuracy: {acc:.4f}, Intra-Class Variance: {var:.4f}")

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(depths, test_accs, marker='o')
plt.xlabel("Number of GCN layers")
plt.ylabel("Test Accuracy")
plt.title("Effect of GCN Depth on Accuracy")

plt.subplot(1,2,2)
plt.plot(depths, intra_class_vars, marker='o', color='red')
plt.xlabel("Number of GCN layers")
plt.ylabel("Intra-Class Variance")
plt.title("Effect of GCN Depth on Hidden Representations")
plt.tight_layout()
plt.show()
```

#### -> 2/4 - layer GCN achieve the highest accuracy (around 0.8), and performance decreases as the network gets deeper. By 16 layers, accuracy drops dramatically.. (Classic sign of over-smoothing in GCNs: as the number of graph convolution layers grows, node embeddings can become indistinguishable from each other..)
#### -> Intra-class variance is smallest at 2 layers and increases with depth. as the GCN goes deeper, node embeddings within the same class become more entangled or flatten out, which leads to larger variance and weaker discriminative power..

#### => A moderate depth (24 layers)  balances the benefits of message passing and the risk of over-smoothing. Beyond a certain point, deeper GCNs typically degrade performance ...
#### -> **Future Directions**: We can employ specific techniques (like residual connections, jumping knowledge, or normalization) to get around this..
